### crop_to_bbox() [UNTESTED]

Crops image to bounding box coordinates, meant for the bounding boxes of lesion markers. It returns the image cropped to the largest area around where the group of markers lie.

#### NOTE: this function is not thouroughly tested, and does not work as intended much of the time


```python
img = Image.open('sample_data/image_1.png').convert("RGB")
display(img)
new_img = enhance_image(img, False)
bbox_list = detect_anno(new_img, False)

if len(bbox_list) > 0:
    cropped_img = crop_to_bbox(img, bbox_list, True)
    display(cropped_img)
else:
    print("no lesions found")
```
